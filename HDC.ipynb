# Heart Disease Detection Classification

```python
# this is for the supervise algorithm
#the type i'm going to use for the training is the CLASSIFICATION type
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#importing the specified modules from sklearn
from sklearn import preprocessing
#below is to implement the logisticRegression
from sklearn.linear_model import LogisticRegression
#this is a function in sklearn that help to split the data(train and test)
from sklearn.model_selection import train_test_split
#to determine the accuracy of the machine learning classifier.
from sklearn.metrics import accuracy_score

%matplotlib inline
```

```python
#df means dataframe
df = pd.read_csv('heart disease classification dataset.csv' , header=0)
```

```python
#must prepare the data in a format that is asked by the classdifer
#building module.
#this is done by encoding

#create one hot encoding of the categorical columns.
data = pd.get_dummies(df,columns = ['age','sex','cp','trestbps','chol','fbs',
                                   'restecg','thalach','exang','oldpeak',
                                   'slope','ca','thal',])
```

```python
#now splitting the dataset into two(the X and Y arrays)

X = data.iloc[:,2:]
```

```python
X.head()
```

```python
#to create an array for the predicted value column

Y = data.iloc[:,1]
```

```python
Y.head()
```

```python
#now below command is use to split the data
#this will create the 4 arrays called X_train,X_test....

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=0)
```

```python
#building the classifier
#using pre-built model from the sklearn

classifier = LogisticRegression(solver='lbfgs',random_state=0)
```

```python
#to tune the classifier

classifier.fit(X_train,Y_train)
```

```python
#to test the algorithm by using the test data generated
#call the predict method

predicted_y = classifier.predict(X_test)
```

```python
predicted_y
```

```python
#just by the way!!!
#to sort out those who have heart disease(yes)
#simply use the below python commands

for x in range(len(predicted_y)):
    if (predicted_y[x] == 'yes'):
        print(x,end="\t")
```

```python
#to test for the accuracy of the prediction
#use the score method on the classifier 

print('Accuracy: {:.2f}'.format(classifier.score(X_test,Y_test)))
```
